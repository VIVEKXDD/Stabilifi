```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 4))

# Create timeline
ax.plot([0, 90], [0, 0], color='black', lw=2)

# Observation Window (T_obs)
obs_rect = patches.Rectangle((0, -0.5), 30, 1, linewidth=2, edgecolor='blue', facecolor='lightblue', alpha=0.5)
ax.add_patch(obs_rect)
plt.text(15, 0.6, 'Observation Window (T_obs)\nDays 1-30\nCalculate Features', ha='center', color='darkblue')

# Performance Window (T_pred)
pred_rect = patches.Rectangle((30, -0.5), 30, 1, linewidth=2, edgecolor='red', facecolor='lightcoral', alpha=0.5)
ax.add_patch(pred_rect)
plt.text(45, 0.6, 'Performance Window (T_pred)\nDays 31-60\nDefine Stress Labels', ha='center', color='darkred')

ax.set_xlim(-5, 95)
ax.set_ylim(-2, 2)
ax.axis('off')
plt.title('Temporal Data Splitting to Prevent Data Leakage', fontsize=14, pad=20)
plt.show()
```
