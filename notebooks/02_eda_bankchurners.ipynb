```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load raw Home Credit / BankChurners dataset [cite: 53, 54]
df_credit = pd.read_csv('../data/raw/BankChurners.csv')
df_credit.head()
```

```python
# Visualize how credit utilization impacts stress (using Attrition as a proxy)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_credit, x='Attrition_Flag', y='Avg_Utilization_Ratio', palette='Set2')
plt.title('Credit Utilization Ratio by Customer Status')
plt.ylabel('Average Utilization Ratio')
plt.xlabel('Customer Status (Proxy for Stress)')
plt.show()
```

```python
# Look at the distribution of revolving balances 
plt.figure(figsize=(10, 6))
sns.histplot(df_credit['Total_Revolving_Bal'], bins=30, kde=True, color='coral')
plt.title('Distribution of Total Revolving Balance')
plt.xlabel('Revolving Balance')
plt.ylabel('Frequency')
plt.show()
```
